# Notebook 2 — Daily exploration and “going home” behaviour

This notebook starts from the cleaned visit-level table produced in Notebook 1
and uses it to describe daily activity patterns and the probability of “going
home” after each visit.

We focus on three objects:

1. The distribution of daily visit counts and distances from home.
2. The next-step outcome after each visit: keep exploring vs go home vs end.
3. The discrete-time hazard of going home after the k-th visit.


```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("default")

# Load the uploaded CSV
path = "data/visit_level_table_000.csv"
visit_table = pd.read_csv(path)

print("Rows in visit_table:", len(visit_table))
visit_table.head()
```

```python
# Parse date + times into full timestamps (optional but handy)
visit_table["date"] = pd.to_datetime(visit_table["date"]).dt.date

visit_table["first_dt"] = pd.to_datetime(
    visit_table["date"].astype(str) + " " + visit_table["first_time"]
)
visit_table["last_dt"] = pd.to_datetime(
    visit_table["date"].astype(str) + " " + visit_table["last_time"]
)

print("Unique persons in this file:", visit_table["person"].unique())
```

## 2. Daily visit counts and distance to HOME

As a first descriptive step we look at:

- how many non-home visits each day contains; and
- how far from HOME these visits typically occur.


```python
# Keep only non-home visits with a positive within-day order
vis = visit_table[
    (visit_table["is_home"] == 0) &
    (visit_table["visit_order_in_day"] > 0)
].copy()

print("Non-home visit events:", len(vis))
print("\nnext_step value counts (non-home):")
print(vis["next_step"].value_counts())
```

```python
# Daily number of non-home visits (max visit_order_in_day per day)
daily_visits = (
    vis.groupby("date")["visit_order_in_day"]
       .max()
       .rename("visits_in_day")
       .reset_index()
       .sort_values("date")
)

daily_visits.head()
daily_stats = daily_visits["visits_in_day"].describe()
daily_stats
```

```python
# --- 2.1 Daily visit counts over time ---
plt.figure(figsize=(10,4))
plt.plot(daily_visits["date"], daily_visits["visits_in_day"],
         marker="o", linewidth=1)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of non-home visits")
plt.title(f"Daily visit counts — user {USER_ID}")
plt.tight_layout()
plt.show()

# --- 2.2 Distance from HOME histogram ---
plt.figure(figsize=(6,4))
plt.hist(vis["dist_home_m"], bins=40)
plt.xlabel("Distance to HOME (m)")
plt.ylabel("Number of visits")
plt.title(f"Distance of non-home visits from HOME — user {USER_ID}")
plt.tight_layout()
plt.show()
```

```python
MAXK = 30        # 只看前 30 次访问
NIGHT_HOUR = 23  # >= 23:00 视作很晚
NEAR_HOME_M = 750

v = vis[vis["visit_order_in_day"] <= MAXK].copy()

# Raw outcome from next_step
v["outcome"] = np.where(
    v["next_step"] == "home", "home",
    np.where(v["next_step"].isin(["sw", "pv", "pn"]), "explore", "end")
)

# Last visit of each day
last_mask = v.groupby("date").cumcount(ascending=False).eq(0)

# Late or near home?
last_time_dt = pd.to_datetime(v["last_time"], format="%H:%M:%S", errors="coerce")
late = last_time_dt.dt.hour.ge(NIGHT_HOUR)
near = v["dist_home_m"].le(NEAR_HOME_M)

# Reclassify terminal 'end' as 'home' if late or near home
v.loc[last_mask & (v["outcome"] == "end") & (late | near), "outcome"] = "home"

print("Outcome counts after reclassification:")
print(v["outcome"].value_counts())
```

```python
# Counts & proportions by visit order k
counts = (
    v.groupby("visit_order_in_day")["outcome"]
     .value_counts()
     .unstack(fill_value=0)
     .reindex(range(1, MAXK+1), fill_value=0)
)
counts["total"] = counts.sum(axis=1)

props = counts.div(counts["total"].where(counts["total"] > 0, np.nan), axis=0)
props_head = props[["explore", "home", "end"]].head(10)
props_head
```

```python
# --- 4.1 Lines: explore vs home ---
plt.figure(figsize=(10,4))
plt.plot(props.index, props["explore"], marker="o", label="Explore")
plt.plot(props.index, props["home"],    marker="o", label="Go home")
plt.xlim(1, MAXK)
plt.ylim(0, 1)
plt.xlabel("Visit order k")
plt.ylabel("Proportion")
plt.title(f"Next-step outcome after the k-th visit — user {USER_ID}")
plt.legend()
plt.tight_layout()
plt.show()

# --- 4.2 Stacked composition ---
plt.figure(figsize=(10,4))
bottom = np.zeros(len(props))
for col, lab in [("explore", "Explore"), ("home", "Home"), ("end", "End/None")]:
    plt.bar(props.index, props[col].fillna(0), bottom=bottom, label=lab)
    bottom += props[col].fillna(0).values

plt.xlim(0, MAXK)
plt.ylim(0, 1)
plt.xlabel("Visit order k")
plt.ylabel("Proportion")
plt.title(f"Next-step composition by visit order — user {USER_ID}")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Per-day max k to define risk set
perday_maxk = v.groupby("date")["visit_order_in_day"].max()

at_risk = pd.Series(
    {k: int((perday_maxk >= k).sum()) for k in range(1, MAXK+1)},
    name="at_risk"
)

home_k = (
    v[v["outcome"] == "home"]
    .groupby("visit_order_in_day")
    .size()
    .reindex(range(1, MAXK+1), fill_value=0)
    .rename("home_k")
)

hazard = (home_k / at_risk.replace(0, np.nan)).rename("hazard")
haz_df = pd.concat([home_k, at_risk, hazard], axis=1)
haz_df.head(10)
```

```python
def wilson_ci(x, n, alpha=0.05):
    if n == 0:
        return (np.nan, np.nan)
    from math import sqrt
    z = 1.959963984540054
    phat = x / n
    denom = 1 + z*z/n
    center = (phat + z*z/(2*n)) / denom
    half = z/denom * sqrt(phat*(1-phat)/n + z*z/(4*n*n))
    return (max(0.0, center-half), min(1.0, center+half))

mask_const = (haz_df.index >= 2) & (haz_df["at_risk"] > 0)
H_num = int(haz_df.loc[mask_const, "home_k"].sum())
H_den = int(haz_df.loc[mask_const, "at_risk"].sum())
h_hat = H_num / H_den
lo, hi = wilson_ci(H_num, H_den)

print(f"Constant hazard ĥ (k≥2): {h_hat:.3f}  95% CI [{lo:.3f}, {hi:.3f}]")
print("Total home events used:", H_num, "  total at risk:", H_den)
```

```python
# === 3. Within-day exploration: when do new places appear? ===

# 3.1 只看非 HOME 的访问事件（HOME 的 visit_order_in_day == 0）
vis_places = visit_table[
    (visit_table["visit_order_in_day"] > 0) &
    (visit_table["is_home"] == 0)
].copy()

MAXK = 30   # 和前面 hazard 一样，限制到 30 步
vis_places = vis_places[vis_places["visit_order_in_day"] <= MAXK].copy()

# 3.2 按日内顺序 k 统计：第 k 次访问是 Pv / Pn 的次数
pv_by_k = (
    vis_places
    .groupby("visit_order_in_day")[["is_pv", "is_pn"]]
    .agg(
        pv_cnt=("is_pv", "sum"),
        pn_cnt=("is_pn", "sum"),
        n=("is_pv", "size")
    )
    .reset_index()
)

pv_by_k["pv_rate"] = pv_by_k["pv_cnt"] / (
    pv_by_k["pv_cnt"] + pv_by_k["pn_cnt"]
).replace(0, np.nan)

overall_pv = vis_places["is_pv"].mean()
print("Overall share of first-time places (Pv) among non-home visits:",
      round(overall_pv, 3))

pv_by_k.head()
```

```python
plt.figure(figsize=(8,4))
plt.plot(
    pv_by_k["visit_order_in_day"],
    pv_by_k["pv_rate"],
    marker="o",
    label="Empirical Pv share"
)

# 整体平均 Pv 比例（水平线）
plt.axhline(
    overall_pv,
    color="gray",
    linestyle="--",
    linewidth=1,
    label="Overall Pv share"
)

plt.ylim(0, 0.4)
plt.xlim(1, MAXK)
plt.xlabel("Visit order k (non-home)")
plt.ylabel("P(visit is Pv | visit at k)")
plt.title(f"Within-day probability of visiting a new place (Pv) — user {USER_ID}")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

import statsmodels.api as sm
import numpy as np

# 3.3 简单 logit：P(Pv_k) = logit^{-1}(α + β (k - mean k))
vis_lr = vis_places.copy()
vis_lr["k_centered"] = (
    vis_lr["visit_order_in_day"] - vis_lr["visit_order_in_day"].mean()
)

X = sm.add_constant(vis_lr["k_centered"])
y = vis_lr["is_pv"]

logit_model = sm.Logit(y, X)
res = logit_model.fit(disp=False)
print(res.summary2().tables[1])

beta0 = res.params["const"]
beta1 = res.params["k_centered"]

# 每多一格访问，去新地点的 odds 乘上多少
odds_ratio = np.exp(beta1)
print(f"Odds ratio for Pv per additional visit: {odds_ratio:.3f}")

# 3.4 画经验值 + logit 平滑曲线
k_grid = np.arange(1, MAXK+1)
k_grid_c = k_grid - vis_lr["visit_order_in_day"].mean()
lin_pred = beta0 + beta1 * k_grid_c
p_hat = 1 / (1 + np.exp(-lin_pred))

plt.figure(figsize=(8,4))
plt.plot(
    pv_by_k["visit_order_in_day"],
    pv_by_k["pv_rate"],
    marker="o",
    label="Empirical Pv share"
)
plt.plot(
    k_grid,
    p_hat,
    linestyle="--",
    label="Logit trend"
)

plt.ylim(0, 0.4)
plt.xlim(1, MAXK)
plt.xlabel("Visit order k (non-home)")
plt.ylabel("P(visit is Pv | visit at k)")
plt.title(f"Exploration within a day (Pv vs k) — user {USER_ID}")
plt.legend()
plt.tight_layout()
plt.show()
```
