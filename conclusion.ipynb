# Results and Discussion

Daily mobility in Geolife is shaped by short home-anchored outings rather than long, continuous paths. Most movement can be read as simple chains that start at HOME, include one or several non-home stops, and return home again, making home–home trips a useful unit for describing behaviour.

Where users differ most is in how these trips end. Only two people show a clear “one-stop errand” pattern with a very high chance of turning back immediately, while the rest display fairly steady return-home probabilities that stay low once a trip has moved past the first stop. The detailed hazard curves look different across individuals, but the shared feature is that return-home chances do not rise with trip length and instead settle into a stable regime after the first few stops.

Exploration shows a similar mix of shared structure and individual variation. Some users become more likely to visit new places (Pv) as trips continue, which means novelty tends to appear in the middle or later part of an outing. Others keep a nearly flat profile, favouring familiar locations regardless of how long the trip lasts. These differences mean that stop order helps explain behaviour but does not collapse all users into a single pattern.

When these pieces—trip length, going-home tendency, and exploration—are viewed together, users fall into a few broad styles rather than one homogeneous group. Some make long, multi-stop trips that revisit known places, some run frequent short errands with quick returns, and some take fewer outings but explore more once they are already out. These contrasts reflect stable behavioural differences between individuals, not noise in the data.
