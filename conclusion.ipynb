# Results and Discussion

The individual-level modelling in Notebook 2 and the cross-user analysis in Notebook 3 together clarify which features of daily mobility exhibit robust population structure and which are strongly heterogeneous across individuals. Rather than revisiting implementation details, this section focuses on the behavioural regularities that emerge from the Geolife sample.

## Population-level behavioural structure

Across users with long Geolife records, daily mobility shows a consistent backbone: people organise their movement into short, home-anchored trips rather than long, continuous excursions. Most mobility unfolds as repeated patterns of “home → one or several non-home stops → home”. This supports treating home–home trips—sequences of non-home visits bracketed by home—as a meaningful structural unit for everyday mobility.

The dynamics of trip termination, however, vary widely across individuals. The discrete-time hazard analysis reveals strong heterogeneity in the “going-home” curve. Some users show a sharp spike in the probability of ending the trip after the first non-home stop; others show only a mild early peak; and a smaller group exhibit their highest hazards later in the trip. What generalises across users is not the location of a peak but the broader tendency for the hazard to settle into a low, stable regime once a trip has passed the initial stops.

Exploration behaviour—measured by first-time visits to places (Pv)—is similarly diverse. For many users, the probability that a stop is a new place increases with stop order, suggesting that exploration is concentrated in the middle and later parts of a trip. For others, the Pv probability is nearly flat, indicating a strong preference for revisiting familiar locations. These patterns confirm that stop order is a meaningful explanatory axis for exploration, while also ruling out a single universal Pv curve, such as a strictly increasing trend.

A multivariate summary of trip counts, stop counts, hazards and exploration rates further shows that users do not form a homogeneous behavioural population. Clustering in this feature space consistently yields three groups: people who take longer, multi-stop trips mostly to familiar places; people whose mobility consists of short, errand-like trips with high early termination; and people who go out infrequently but, when they do, stay out longer and explore new places more intensely. Their separation in principal-component space indicates that this heterogeneity is systematic rather than noise.

Taken together, these cross-user results depict daily mobility as a system with a shared structure and substantial individual differentiation. The shared structure lies in the centrality of home–home trips and the stabilisation of ending probabilities once a trip is underway. The differentiation lies in how strongly early stops matter, how exploration unfolds over the course of a trip, and how each user balances novelty and routine.
